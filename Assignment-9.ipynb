```python
#Q1: Sum of Three Numbers
#Write a function `sum_three(a, b, c)` that returns the sum of the three numbers.
#Example:
#sum_three(3, 5, 7) ➞ 15

def summ(a,b,c):
    return(a+b+c)
def sum1():
    s=summ(3,5,7)
    print(f'sum of the three Number:{s} ')
sum1()
```

```python
#Q2: Area of a Circle
#Write a function `area_circle(radius)` that returns the area of a circle. Use π = 3.14159
#Example:
#area_circle(7) ➞ 153.9381
def area_circle(radius):
    pi=3.14159
    return pi * radius*radius  
    
def result():
    radius=7
    area=area_circle(radius)
    print(f"Area of the circle with radius {radius}: {area}")

result()
```

```python
#Q3: Check Even or Odd
#Write a function `check_even_odd(num)` that prints whether a number is Even or Odd.
#Example:
#check_even_odd(5) ➞ &quot;Odd&quot;
#check_even_odd(8) ➞ &quot;Even&quot;

def check_even_odd(num):
    if num%2==0:
        return('even')
    else:
        return('odd')

print(check_even_odd(5))
print(check_even_odd(8))
```

```python
#Q4: Calculate BMI
#Write a function `calculate_bmi(weight, height)` that returns BMI value using the formula:
#BMI = weight / (height ** 2)
#Example:
#calculate_bmi(70, 1.75) ➞ 22.86

def calculate_bmi(weight, height):
    b=weight / (height ** 2)
    return round(b,2)

def calculate():
    bmi=calculate_bmi(70, 1.75)
    print(f'Total : {bmi}')

calculate()
```

```python
#Q6: Find Maximum of Two Numbers
#Write a function `find_max(a, b)` that returns the greater of the two numbers.
#Example:
#find_max(4, 9) ➞ 9

def find_max(a, b):
    return max(a,b)

print(find_max(1130,250))
```

```python
#Q7: Power Function
#Write a function `power(base, exponent)` that returns base raised to the power exponent.
#Example:
#power(2, 3) ➞ 8

def power(base, exponent):
    return((base) ** exponent)

print(power(2,5))
```

```python
#Q8: Calculate Simple Interest
#Write a function `simple_interest(principal, rate, time)` that returns the simple interest.
#Formula: (P × R × T) / 100
#Example:
#simple_interest(1000, 5, 2) ➞ 100.0

def simple_interest(principal, rate, time):
    return((principal * rate * time)/100)

def Calculate():
    s=simple_interest(1000, 5, 2)
    principal=1000
    rate=5
    time=2
    print(f'Total {principal}*{rate}*{time}/100  : {s}')

Calculate()
```

```python
#Q9: Check Leap Year
#Write a function `is_leap_year(year)` that returns True if the year is a leap year, otherwise##
#False.
#Example:
#is_leap_year(2020) ➞ True
#is_leap_year(2023) ➞ False

def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
print(is_leap_year(2020))
print(is_leap_year(2023))
print(is_leap_year(2016))
```
