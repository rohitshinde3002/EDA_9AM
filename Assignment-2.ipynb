```python
# wap ask the user take two number print subtraction multiplication and division
# ans: the subtraction of 100 and 200 is -100
#      the multiplication of 100 and 200 is 20000
#      the division of 100 and 200 is 0.5

a=eval(input('enter value:'))
b=eval(input('enter value:'))
sub=(a-b)
mul=(a*b)
div=(a/b)
print(f'the sub of {a} and {b} is: {sub}')
print(f'the mul of {a} and {b} is: {mul}')
print(f'the div of {a} and {b} is: {div}')
print('  The sub of {} and {} is: {}'.format(a,b,sub))
print('  The mul of {} and {} is: {}'.format(a,b,mul))
print('  The div of {} and {} is: {}'.format(a,b,div))

```

```python
# wap ask the user take the radius find the area of circle
    # pi=3.14
    # r=100
   # area_of_circle=pi*r*r

pi=3.14
r=eval(input('Enter the no:'))
area_of_circle=pi*r*r
print(f'area of circle: {pi}*{r}*{r} = {area_of_circle}')
print('area of circle: {}*{}*{} = {}'.format(pi,r,r,area_of_circle))
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h

b=eval(input('enter the No:'))
h=eval(input('enter the No:'))
area_of_traingle=(0.5 * b * h)
print(f'area of traingle: 0.5*{b}*{h} = {area_of_traingle}')
print('area of traingle: 0.5*{}*{} = {}'.format(b,h,area_of_traingle))
```

```python
# wap ask the user take 3 numbers find the average
a=eval(input('enter value:'))
b=eval(input('enter value:'))
c=eval(input('enter value:'))
avg=(a+b+c)/3
print(f'average: {a}+{b}+{c} = {avg}')
print('average: {}+{}+{} = {}'.format(a,b,c,avg))
```

```python
# wap ask the user take the bill amount and tip amount calculate total bill
bill_amount=eval(input('Enter the no:'))
tip_amount=eval(input('Enter the no:'))
total=(bill_amount+tip_amount)
print(f'Total amount: {bill_amount}+{tip_amount} = {total}')
print('Total amount: {}+{} = {}'.format(bill_amount,tip_amount,total))
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount
#      tax_amount= salary*tax_per/100

salary=eval(input('Enter the no:'))
tax_per=eval(input('Enter the no:'))
tax_amount= salary*tax_per/100
print(f'total: {salary} * {tax_per} / 100 = {tax_amount}')
print('total: {} * {} / 100 = {}'.format(salary,tax_per,tax_amount))
```
