```python
# 1.Write a program to check if a number entered by the user is even or odd.

try:
    num=eval(input('Enter the number: '))
    if num%2==0:
        print(f'the {num} is even:')
    else:
        print(f'the {num} is odd:')

except Exception as e:
    print(e)
    
```

```python
# 2.Write a program to check whether a number is positive, negative, or zero

try:
    num1=eval(input('Enter the number: '))
    if num1>0:
        print(f'The {num1} is Positive')
    elif num1<0:
        print(f'The {num1} is Negative')        
    else:
        print(f'The {num1} is Zero')
except Exception as e:
    print(e)
```

```python
# 3.Write a program that takes two numbers and prints the larger one.

try:
    n1=eval(input('Enter the number: '))
    n2=eval(input('Enter the number: '))
    if n1>n2:
        print(f'The larger number: {n1}')
    elif n2>n1:
        print(f'The larger number: {n2}')        
    else:
        print(f'The both are equal {n1}={n2}')
except Exception as e:
    print(e)
```

```python
# 4.Ask the user to enter their age and check if they are eligible to vote (age >= 18).

try:
    age=int(input('Enter the age: '))
    if age>=18:
        print(f'you are eligible to vote : {age} ')
    else:
        print(f'you are not eligible to vote : {age} ')

except Exception as e:
    print(e)
```

```python
# 5.Check if a number is divisible by both 5 and 11.

try:
    num2 = int(input('Enter the number: '))

    if num2 % 5 == 0 and num2 % 11 == 0:  
        print("The number is divisible by both 5 and 11.")
    else:
        print("The number is not divisible by both 5 and 11.")

except Exception as e:
    print(e)
```

```python
# 6.Check if a number is divisible by both 3 and 5.

try:
    num=int(input('Enter the no: '))
    if num%3==0 and num%5==0:       
        print("The number is divisible by both 3 and 5.")
    else:
        print("The number is not divisible by both 3 and 5.")

except Exception as e:
    print(e)
```

```python
# 7.Take an amount as input. 
#   If it's a multiple of 100 and less than balance, print "Withdrawal Successful", else print reason for failure.

try:
    balance = 5000  
    a = float(input("Enter the amount: "))

    if a % 100 == 0 and a <= balance:
        print("Withdrawal Successful")
    else:
        print("Withdrawal Failed")

except Exception as e:
    print(e)

```
