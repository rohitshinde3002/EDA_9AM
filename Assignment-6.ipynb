```python
# 1. Ramesh’s basic salary is input through the keyboard. His dearness allowance is 40% of basic salary, and
#    house rent allowance is 20% of basic salary. Write a program to calculate his gross salary.

salary=eval(input('Enter the salary:'))
da=40*salary
hra=20*salary
total=salary+da+hra
t1=round(total,2)
print(f'gross salary is: {salary}+{da}+{hra}={t1}')
```

```python
#2. The distance between two cities (in km.) is input through the keyboard. Write a program to convert and
#    print this distance in meters, feet, inches and centimeters.

km=eval(input('Enter the km:'))
distance_meters=km*1000
distance_feet=km*3280
distance_inches=km*39370
distance_centimeters=km*100000
print(f'distance_meters=   {km}*1000={distance_meters}')
print(f'distance_feet=     {km}*3280={distance_feet}')
print(f'distance_inches=   {km}*39370={distance_inches}')
print(f'distance_centimeters={km}*100000={distance_centimeters}')
```

```python
#3. If the marks obtained by a student in five different subjects are input through the keyboard, find out the
#   aggregate marks and percentage marks obtained by the student. Assume that the maximum marks that
#   can be obtained by a student in each subject is 100.

english=eval(input('Enter the marks:'))
hindi=eval(input('Enter the marks:'))
marathi=eval(input('Enter the marks:'))
maths=eval(input('Enter the marks:'))
physics=eval(input('Enter the marks:'))
aggregate_marks=english+hindi+marathi+maths+physics
print(f'aggregate marks={english}+{hindi}+{marathi}+{maths}+{physics}={aggregate_marks}')
total_marks=100*5
percentage_marks = (aggregate_marks / total_marks) * 100
p1=round(percentage_marks,3)
print(f'percentage marks:({aggregate_marks}/{total_marks})*{100}={p1}')

```

```python
#4. Temperature of a city in Fahrenheit degrees is input through the keyboard. Write a program to convert this
#     temperature into Centigrade degrees.

f=eval(input('Fahrenheit:'))
c=5/9*(f-32)
c1=round(c,2)
print(f'Centigrade:{c1}')
```

```python
#5. The length & breadth of a rectangle and radius of a circle are input through the keyboard. Write a program
#    to calculate the area & perimeter of the rectangle, and the area & circumference of the circle.

l=float(input('Enter the Lenght:'))
b=float(input('Enter the breadth:'))
print("--- rectangle ---")
area = l * b
Perimeter = 2 * (l + b)
print(f'area:{l} * {b} = {area}')
print(f'Perimeter:{2} * (l+b) = {Perimeter}')

# area and circumference of a circle,- formulas

π = 3.14
radius=float(input('Enter the radius'))

area1 = π * radius*radius
Circumference = 2 * π * radius
a=round(area1,2)
c=round(Circumference,2)

print("--- Circle ---")
print(f'area1:{π} * {radius}*{radius}={a}')
print(f'Circumference:{2} * {π} * {radius}={c}')

```

```python
# 7. If a five-digit number is input through the keyboard, write a program to reverse the number.

num = input("Enter a number: ")
reversed_num = num[::-1]

print(f"Reversed number: {reversed_num}")
```

```python
# 8. If a four-digit number is input through the keyboard, write a program to obtain the sum of the first and last
#    digit of this number.

num = int(input("Enter a four-digit number: "))
first = num // 10000
last = num % 10
result = first + last

print(f"Sum of the first and last digit: {result}")
```

```python
#9. In a town, the percentage of men is 52. The percentage of total literacy is 48. If total percentage of literate
#    men is 35 of the total population, write a program to find the total number of illiterate men and women if
#    the population of the town is 80,000.

population = 80000

men = 0.52 * population
women = population - men

literate_men = 0.35 * population
literate_total = 0.48 * population

illiterate_men = men - literate_men
literate_women = literate_total - literate_men
illiterate_women = women - literate_women

print("Illiterate men:", int(illiterate_men))
print("Illiterate women:", int(illiterate_women))
```
