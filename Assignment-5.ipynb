```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h
import random
b=random.randint(1,100)
h=random.randint(1,100)
area_of_traingle=0.5*b*h
print(f'area_of_traingle:{0.5}*{b}*{h}= {area_of_traingle}')
```

```python
# wap ask the user take 3 numbers find the average
import random
n1=random.randint(1,100)
n2=random.randint(1,100)
n3=random.randint(1,100)
avg=(n1+n2+n3)/3
avg1=round(avg,3)
print(f'average: {n1}+{n2}+{n3} = {avg1}')
```
