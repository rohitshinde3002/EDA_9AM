```python
# wap ask the user take two number print subtraction multiplication and division

def sub(a, b):
    return a - b

def mult(a, b):
    return a * b

def div(a, b):
    if b != 0:
        return a / b
    else:
        return "0-division"

def operations():
    n1=int(input('Enter the no:'))
    n2=int(input('Enter the no:'))

    print('sub :',n1 - n2)
    print('mult:',n1 * n2)
    print('div :',n1 / n2)

operations()
  
```

```python
# wap ask the user take the radius find the area of circle
    # pi=method use 
    # r=100
   # area_of_circle=pi*r*r
def pi():
    return (3.14)
def area_of_circle():
    r=eval(input('Enter the no:'))
    pi=3.14
    circle=pi*r*r
    c1=round(circle,3)
    print(f'area of circle: {pi}*{r}*{r}={c1}')

area_of_circle()
```

```python
# wap ask the user take height and breadth find the area of traingle
#       area_of_traingle=0.5*b*h

def triangle_area(b, h):
    return (0.5 * b * h)
def area_of_traingle():
    b=eval(input('Enter the no:'))
    h=eval(input('Enter the no:'))
    
    area=triangle_area(b, h)

    print(f' area-of-traingle=0.5*{b}*{h} : {area}')
    
area_of_traingle()
```

```python
# wap ask the user take 3 numbers find the average
def avg1(a,b,c):
    return((a+b+c)/3)
def avg():
    n1=int(input('Enter the no:'))
    n2=int(input('Enter the no:'))
    n3=int(input('Enter the no:'))
    f=avg1(n1,n2,n3)
    print(f'Total average- {n1}+{n2}+{n3}/3: {f}')

avg()
```

```python
# wap ask the user take the bill amount and tip amount calculate total bill

def bill(a,b):
    return(a+b)
def total_bill():
    n1=int(input("enter the bill amount:"))
    n2=int(input("enter the tip amount: "))
    total=bill(n1,n2)
    print(f'Total bill:{n1}+{n2}={total}')

total_bill()
    
```

```python
# wap ask the user take the salary, percetage of tax calculate tax amount

def tax1(a,b):
    return(a*b/100)
def tax():
    salary=int(input('enter the amount: '))
    tax_per=int(input('enter the tax no: '))
    
    tax_amount=tax1(salary,tax_per)
    print(f'total_amount= {salary}*{tax_per}/100 = {tax_amount}')

tax()
```

```python
# wap ask the user take the value in dollar convert into rupees
#         1$=80rs
def val(dollar):
    return(dollar * 80)

def value():
    dollar=float(input('Enter the dollar:'))
    rupees = val(dollar)
    print(f'{dollar} *80 convert into = {rupees}')

value()
```
