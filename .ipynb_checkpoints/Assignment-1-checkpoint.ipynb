**2. Calculate simple interest**

simple_interest = (principal * rate * time) / 100


Ask user for principal, rate, and time


Display the result

```python
principal=100
rate=200
time=2
simple_interest = (principal * rate * time) / 100
simple_interest
print('simple interest  ({} * {} * {})/100 = {}'.format(principal,rate,time,simple_interest))
print(f'simple interest  ({principal} * {rate} * {time})/100 = {simple_interest}')
```

**3. Convert temperature from Celsius to Fahrenheit**

fahrenheit = (celsius * 9/5) + 32


Ask user for temperature in Celsius


Display in Fahrenheit

```python
celsius=25
fahrenheit = (celsius * 9/5) + 32
print('Temperature in Fahrenheit: ({} * 9/5) + 32  = {}'.format(celsius,fahrenheit))
print(f'Temperature in Fahrenheit: ({celsius} * 9/5) + 32  = {fahrenheit}')
```

**4. Convert kilometers to miles**

miles = kilometers * 0.621371


Ask user for distance in kilometers


Display miles


```python
kilometers=10
miles = kilometers * 0.621371
miles
print('distance_in_kilometers: {} * 0.621371 = {}'.format(kilometers,miles))
print(f'distance_in_kilometers: {kilometers} * 0.621371 = {miles}')
```

**7. Calculate the perimeter of a rectangle**

**8. Find the average of three numbers**

average = (num1 + num2 + num3) / 3

**10. Calculate discount price**

discounted_price = original_price - (original_price * discount / 100)


Ask user for original_price and discount percentage


Display final price

```python
original_price=100
discount=20
discounted_price = original_price - (original_price * discount / 100)
print('final price : {} - ({} * {} / 100) = {}'.format(original_price,original_price,discount,discounted_price))
print(f'final price : {original_price} - ({original_price} * {discount} / 100) = {discounted_price}')
```
